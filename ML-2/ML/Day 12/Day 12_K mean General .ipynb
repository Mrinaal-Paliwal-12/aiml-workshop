```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
#make blobs are used to create data samples 
X,y_true = make_blobs(n_samples=300,
                      centers=4,cluster_std=0.5,random_state=0)
plt.scatter(X[:,0],X[:,1],s=50);
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
```

```python
y_kmeans
```

```python
plt.scatter(X[:,0],X[:,1],c=y_kmeans,s=50,cmap='viridis') #for points
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.5)#for centers 
```

```python
from sklearn.datasets import load_digits #inbuilt pakage in sklearn
digits=load_digits() #this will load the dataset
digits.data.shape
```

```python
digits.data
```

```python
kmeans=KMeans(n_clusters=10,random_state=42)
clusters=kmeans.fit_predict(digits.data) #only the x values no y values becoz its unsupervise
kmeans.cluster_centers_.shape
```

```python
clusters
```

```python
kmeans.cluster_centers_
```

```python
fig,ax=plt.subplots(2,5,figsize=(8,3))
centers=kmeans.cluster_centers_.reshape(10,8,8)
for axi,center in zip(ax.flat,centers):
    axi.set(xticks=[],yticks=[])
    axi.imshow(center,interpolation='nearest',cmap=plt.cm.binary)
```

```python
#import seaborn as sns;
#sns.set()# for plot styling
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(digits.target,clusters)
mat
```

# K Mean on IRIS DATASET

```python
df=pd.read_csv("D:\\vk\\TRIM 3\\ML\\DATASET\\iris.csv")
```

```python

df
```

```python
X=df[['Sepal length','Sepal width','Petal length','Petal width']].values
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
```

```python
print(y_kmeans)
```

```python
kmeans.cluster_centers_
```

```python
n_cluster=3
silhouette_avg=silhouette_score(X,y_kmeans)
print("For n_cluster= ",n_cluster,"average silhouette_score is :",silhouette_avg)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_kmeans,s=50,cmap='rainbow')
```

```python
Error =[]
for i in range (2,11):
     kmeans = KMeans(n_clusters = i).fit(X)
#    kmeans.fit(x)
     Error.append(kmeans.inertia_)

plt.plot(range(2, 11), Error)
plt.title( 'Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt. show()
```
